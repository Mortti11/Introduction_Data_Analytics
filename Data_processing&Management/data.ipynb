<b>Use web scraping (e.g. BeautifulSoup), and use it to get the
following info from the Rovaniemi Wikipedia –page <b>

```python
url = "https://en.wikipedia.org/wiki/Rovaniemi"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
city = soup.find("h1", id="firstHeading").text
city
```

<b>Get the coordinates of Rovaniemi (upper right corner)<b>

```python
# Finding the coordinates of the city
coordinates = soup.find("span", class_="geo-dec").text
coordinates
```

<b>Get the nickname of Rovaniemi (under coat of arms)<b>

```python
# Finding the Nickname of the city
nickname = soup.find("div", class_= 'ib-settlement-nickname nickname').text
nickname
```

<b>Get the total population<b>

```python
population = soup.find_all("td", class_="infobox-data")

for i in population:
    if "," in i.text and i.text.replace(",", "").isdigit():  
        population = i.text.strip()
        break

population
```

```python
# I tried to row.find("th") to look for <th> tags labeled "Population".
```

<b> Extra task: Get the total area<b>

```python
area = soup.find_all("td", class_="infobox-data")

for i in area:
    if "km2" in i.text: 
        area = i.text.split()[0].replace(",", "")
        break
float(area)
```

```python
unfix = pd.read_csv('Datasets/temperatures_unfixed.csv')
unfix.head(20)
```

```python
unfix[unfix['Date'] == '1.4.2006']
```

```python
# I have to copy the data to keep the original data
AveSummary = unfix.copy()
# I will use lambda function to find most common summary for each day
AveSummary["most_freq_summary"] = (AveSummary.groupby("Date")["Summary"].transform(lambda x: x.mode()[0]))
AveSummary.head(10)
```

```python
# I have drop the duplicated days from the data and also the Summary column and Hour 
AveSummary = AveSummary.drop_duplicates(subset='Date', keep='first')
AveSummary = AveSummary.drop(['Summary', 'Hour'], axis=1)
```

```python
AveSummary.head(10)
```

```python
# Saving the data
AveSummary.to_csv('AveSummary.csv', index=False)
```

```python
# Grouping the data by date and calculating the average temperature for each day
average = unfix.groupby('Date')['Temperature (C)'].mean().reset_index()
```

```python
# Save the data to a new csv file
average.to_csv('average.csv', index=False)
```
