<b>Load the "simple.json" –file without pandas, and try to get the
average price with plain Python<b>

```python
with open("Datasets/simple.json", "r") as f:
    data = json.load(f)
data
```

```python
# Normalizing the json data
data = pd.json_normalize(data, sep='_')
data
```

```python
# Getting the average price without external libraries
prices = data['price']
total  = 0
for price in prices:
    total = sum(prices)
average = total / len(prices)
average
```
