### Functions and lambdas allow us to extend the operations we wish  to columns and rows in pandas. 

```python
data = pd.read_csv('Datasets/loans.csv')
data.head()
```

```python
# Using  lambda for categorizing the Annual Income
data[' Group range'] = data['Annual Income'].apply(
    lambda x: (
        "$25k or less" if x <= 25000 else
        "$25k-$50k" if 25001 <= x <= 50000 else
        "$50k-$100k" if 50001 <= x <= 100000 else
        "$100k-$200k" if 100001 <= x <= 200000 else
        "$200k+" if x > 200000 else
        "Unknown"
    )
)

# I got this syntax from the internet :https://stackoverflow.com/questions/70602832/how-to-categorize-value-using-apply-function-and-lambda
"""
this is the code that I used to categorize the age group
df.age = df.age.astype(int)
df['category'] = df['age'].apply(lambda x: 'mature' if x > 20 else 'kids' if x < 20 else 'youth')
"""
```
